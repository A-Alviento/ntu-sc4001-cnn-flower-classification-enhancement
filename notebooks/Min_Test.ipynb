```python
import matplotlib.pyplot as plt
import torchvision.datasets as datasets

from torch.utils.data import DataLoader
```

```python
train_dataset = datasets.Flowers102(root='../data', split='train', download=True)
test_dataset = datasets.Flowers102(root='../data', split='test', download=True)
val_dataset = datasets.Flowers102(root='../data', split='val', download=True)

train_len = len(train_dataset)
test_len = len(test_dataset)
val_len = len(val_dataset)

train_len, test_len, val_len
```

```python
train_loader = DataLoader(train_dataset, batch_size=128, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=128, shuffle=True)
```

## The train and validation datasets are class-balanced but the train-test split is unconventional (1020:6149).
## We can't choose to use the test dataset as the new train dataset since there is a class imbalance in the test dataset.
## Perhaps data augmentation will help increase the exisiting train dataset size.

```python
fig, axs = plt.subplots(1, 3, figsize=(20,5))

axs[0].hist(list(c for _, c in train_dataset))
axs[0].set_title("Train Dataset Classes")
axs[0].set_xlabel('Classes')
axs[0].set_ylabel('Count')

axs[1].hist(list(c for _, c in val_dataset))
axs[1].set_title("Validation Dataset Classes")
axs[1].set_xlabel('Classes')
axs[1].set_ylabel('Count')

axs[2].hist(list(c for _, c in test_dataset))
axs[2].set_title("Test Dataset Classes")
axs[2].set_xlabel('Classes')
axs[2].set_ylabel('Count')

plt.show()
```
